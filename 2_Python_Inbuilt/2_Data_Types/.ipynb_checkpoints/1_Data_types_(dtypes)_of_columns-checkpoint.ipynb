### See the data types of each column in your dataframe using the .dtypes property.
  
  - Notes that character / string columns appear as ‘object’ datatypes.

### To change the datatype of a specific column, use the .astype( ) function.
  
  - For example, to see the ‘Item Code’ column as a string, use:

```python
df['Item Code'].astype(str)
```
