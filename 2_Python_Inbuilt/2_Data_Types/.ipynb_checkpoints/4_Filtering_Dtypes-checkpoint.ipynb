```python
import pandas as pd
```

```python
data = {
    'intcol': [1,2,3,4,5],
    'floatcol': [1.1, 1.2, 1.3, 1.4, 1.5],
    'objcol': ['A','B','C','D','E'],
    'boolcol': [True, True, False, False, True]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.select_dtypes(include=['bool','int64'])
```

```python
df.select_dtypes(exclude=['bool','int64'])
```

```python
df.select_dtypes(include=['O'])
```

### Concatenate : Train & Test
